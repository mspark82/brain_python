# 7장 함수로 코드 간추리기

```python
# 가변 매개변수
def merge_string(*text_list):
    result = ''
    for s in text_list:
        result = result + s
    return result

merge_string('a','b','c')

# 딕셔너리 기반 매개변수
def print_team(**players):
    for k in players.keys():
        print('{0} = {1}'.format(k, players[k]))

print_team(카시아스='GK', 호날두='FM',알론소='MF', 페페='DF')

# 일반 매개변수 + 가변 매개변수
def print_args(argc, *argv):
    for i in range(argc):
        print(argv[i])

print_args(3, 'argv1', 'argv2', 'argv3')

# 가변 매개변수 + 일반 매개변수
def print_args(*argv, argc):
    for i in range(argc):
        print(argv[i])

print_args('argv1', 'argv2', 'argv3',argc=3)   # 가변 매개변수 뒤에 정의된 일반 매개변수는 반드시 키워드 매개변수로 호출
```

```python
# 재귀함수 - 무한루프
def no_idea():
    print('나는 아무 생각이 없다.')
    print('왜냐하면')
    no_idea()

no_idea()
```

```python
# 연습문제 3
def average(*numbers):
    result_sum = 0
    for i in numbers:
        result_sum = result_sum + i
    
    return result_sum / len(numbers)

average(1,2,3,4,5)
```

```python
# 연습문제 5
def print_star(num):
    for i in range(num, 0, -1):
        for j in range(i):
            print('*', end ='')
        print()

print_star(10)
```
